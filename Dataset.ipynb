```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with columns 'alcohol' and 'quality'
plt.figure(figsize=(10, 6))

# Create a histogram of 'alcohol' for each 'quality' category
for quality_value in sorted(df['quality'].unique()):
    subset = df[df['quality'] == quality_value]
    plt.hist(subset['alcohol'], bins=20, alpha=0.5, label=f'Quality {quality_value}')

plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Histogram of Alcohol Content Against Quality')
plt.legend()
plt.grid(True)
plt.show()
```
